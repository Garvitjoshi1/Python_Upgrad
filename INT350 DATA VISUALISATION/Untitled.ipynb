## `Line Chart:`

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales = [5000, 6000, 7000, 8000, 9000, 10000, 11000, 12000, 13000, 14000, 15000, 16000]

plt.plot(months, sales, marker='o')
plt.xlabel('Month')
plt.ylabel('Sales (in USD)')
plt.title('Monthly Sales Data')
plt.grid(True)
plt.show()
```

## ` Bar Graph:`

```python
plt.bar(months, sales, color='blue')
plt.xlabel('Month')
plt.ylabel('Sales (in USD)')
plt.title('Monthly Sales Data')
plt.xticks(rotation=45)
plt.show()
```

## `Box Plot:`

```python
import seaborn as sns

sns.boxplot(y=sales)
plt.ylabel('Sales (in USD)')
plt.title('Sales Distribution')
plt.show()
```

## `Stacked bar graph:`

```python
import matplotlib.pyplot as plt

regions = ['North', 'South', 'West']
electronics_sales = [5000, 7000, 6000]
apparel_sales = [3000, 4000, 3500]

plt.bar(regions, electronics_sales, label='Electronics', color='blue')
plt.bar(regions, apparel_sales, bottom=electronics_sales, label='Apparel', color='orange')

plt.xlabel('Region')
plt.ylabel('Sales (in USD)')
plt.title('Sales Data by Region and Product Category')
plt.legend()
plt.show()
```
