### Data Cleaning

```python
df = pd.read_csv('bank_data.csv', skiprows = 2)   #skip_blank_lines,skipfooter,skipinitialspace
df.head()
# The skiprows parameter is used in data processing libraries like Pandas to skip a specified number of initial rows when reading files.
```

### Dropping column

```python
df.drop("customerid",axis=1, inplace=True)
# "inplace" apply changes directly to the existing data structure without creating a new one
df.head()
# axis=0 refers to operations performed along rows (vertically), and axis=1 refers to operations performed along columns (horizontally) in a DataFrame or array.
```

### Dividing jobedu in job and education

```python
df['Job'] = df.jobedu.apply(lambda x: x.split(",")[0])
df.head()
```

```python
df['Education'] = df.jobedu.apply(lambda x: x.split(",")[1])
df.head()
```

### Dropping the jobedu

```python
df.drop("jobedu",axis=1, inplace=True)
df.head()
# "drop" is used to remove rows or columns from a DataFrame
```
